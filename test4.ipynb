```python
 What is the difference between a list, tuple, and set in Python? Provide examples.
```

```python
In Python, Lists, Sets and Tuples store collections but differ in behavior.
Lists are ordered, mutable and allow duplicates, suitable for dynamic data.
Sets are unordered, mutable and unique, 
while Tuples are ordered, immutable and allow duplicates, ideal for fixed data.


```

```python
 How does Python manage memory? Explain the concept of garbage collection.
```

```python
Python uses a hybrid approach of reference counting and generational garbage collection to manage memory,
automatically reclaiming memory occupied by objects that are no longer in use, preventing memory leaks and
allowing developers to focus on code rather than memory management. 
```

```python
Explain the difference between a population and a sample. Why do we use sampling?
```

```python
A population refers to the entire group of individuals or items that a researcher is interested in studying, 
while a sample is a smaller subset of that population selected to gather data from, allowing researchers 
to make inferences about the larger population based on the characteristics of the sample.
```

```python
If a number is divisible by 3, print "Fizz"
```

```python
num = int(input("Enter a number: "))

if num % 3 == 0:
    print("Fizz")
else:
    print(num)
```

```python
If a number is divisible by 5, print "Buzz".
```

```python
num = int(input("Enter a number: "))

if num % 5 == 0:
    print("buzz")
else:
    print(num)
```

```python
If a number is divisible by both 3 and 5, print "FizzBuzz".
```

```python
num = int(input("Enter a number: "))

if num % 3 == 0 and num % 5 == 0:
    print("FizzBuzz")
elif num % 3 == 0:
    print("Fizz")
elif num % 5 == 0:
    print("Buzz")
else:
    print(num)
```
