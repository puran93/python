```python
training_set = training_set.iloc[:,1:2].values
test_set = test_set.iloc[:, 1:2].values
training_set
```

```python
sc = MinMaxScaler()
training_set = sc.fit_transform(training_set)
training_set
```

```python
X_train = training_set[:59]
y_train = training_set[1:60]
```

```python
print(f"Before Rehshape {X_train.shape}")
X_train = np.reshape(X_train, (59,1,1))
print(f"After Rehshape {X_train.shape}")
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
model = Sequential()
```

```python
model.add(LSTM(units = 4, activation = 'sigmoid', input_shape = (None, 1)))
```

```python
model.add(Dense(units = 1))
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model.fit(X_train, y_train, batch_size = 16, epochs = 50)
```

```python
real_stock_price = test_set
inputs = real_stock_price
inputs = sc.transform(test_set)
inputs = np.reshape(inputs,(inputs.shape[0],1,1))
predicted_stock_price = model.predict(inputs)
```

```python
predicted_stock_price = sc.inverse_transform(predicted_stock_price)


plt.plot(real_stock_price , color = 'red' , label = 'Real Google Stock')
plt.plot(predicted_stock_price , color = 'blue' , label = 'Predicted Google Stock')
plt.title('Google Stock Prediction')
plt.xlabel('time')
plt.ylabel('Google Stock')
plt.legend()
plt.show()
```
